```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

train_data = pd.read_csv(r"D:\Projects\House-Prices-ML\train.csv")# your path here
test_data = pd.read_csv(r"D:\Projects\House-Prices-ML\test.csv")# your path here

y = train_data["SalePrice"]

features = [
    "OverallQual",  # Overall quality of the house
    "GrLivArea",    # Above ground living area square feet
    "Neighborhood", # Location within Ames city limits
    "TotalBsmtSF",  # Total square feet of basement area
    "YearBuilt",    # Original construction date
    "YearRemodAdd", # Remodel date (same as construction date if no remodeling)
    "GarageCars",   # Size of garage in car capacity
    "GarageArea",   # Size of garage in square feet
    "FullBath",     # Full bathrooms above grade
    "1stFlrSF",     # First floor square feet
    "KitchenQual",  # Kitchen quality
    "TotRmsAbvGrd", # Total rooms above grade (excluding bathrooms)
    "Fireplaces",   # Number of fireplaces
    "LotArea",      # Lot size in square feet
    "BsmtFinSF1"    # Type 1 finished square feet in basement
]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

X, X_test = X.align(X_test, join='left', axis=1, fill_value=0)

model = RandomForestRegressor(n_estimators=1000, max_depth=5, random_state=0)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'Id': test_data.Id, 'SalePrice': predictions})

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.33, random_state=0)
valid_predictions = model.predict(X_valid)
log_valid_predictions = np.log(valid_predictions)
log_y_valid = np.log(y_valid)
rmse = np.sqrt(mean_squared_error(log_y_valid, log_valid_predictions))
print(f"RMSE: {rmse}")

output.head(10)
```

```python
import pandas as pd
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

train_data = pd.read_csv(r"D:\Projects\House-Prices-ML\train.csv")  # your path here
test_data = pd.read_csv(r"D:\Projects\House-Prices-ML\test.csv")    # your path here

y = train_data["SalePrice"]

features = [
    "OverallQual",  # Overall quality of the house
    "GrLivArea",    # Above ground living area square feet
    "Neighborhood", # Location within Ames city limits
    "TotalBsmtSF",  # Total square feet of basement area
    "YearBuilt",    # Original construction date
    "YearRemodAdd", # Remodel date (same as construction date if no remodeling)
    "GarageCars",   # Size of garage in car capacity
    "GarageArea",   # Size of garage in square feet
    "FullBath",     # Full bathrooms above grade
    "1stFlrSF",     # First floor square feet
    "KitchenQual",  # Kitchen quality
    "TotRmsAbvGrd", # Total rooms above grade (excluding bathrooms)
    "Fireplaces",   # Number of fireplaces
    "LotArea",      # Lot size in square feet
    "BsmtFinSF1"    # Type 1 finished square feet in basement
]

X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

X, X_test = X.align(X_test, join='left', axis=1, fill_value=0)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.33, random_state=0)

model = HistGradientBoostingRegressor(max_iter=1000, max_depth=5, learning_rate=0.01 ,random_state=0)
model.fit(X_train, y_train)
valid_predictions = model.predict(X_valid)

log_valid_predictions = np.log(valid_predictions)
log_y_valid = np.log(y_valid)
rmse = np.sqrt(mean_squared_error(log_y_valid, log_valid_predictions))
print(f"RMSE: {rmse}")

test_predictions = model.predict(X_test)
output = pd.DataFrame({'Id': test_data.Id, 'SalePrice': test_predictions})

output.head(10)
```
